```python
# Importation des bibliothèques nécessaires
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import joblib
import os
from datetime import datetime, timedelta

# Configuration de l'affichage des graphiques
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12

# 1. Récupération des données Bitcoin
print("Téléchargement des données Bitcoin...")
btc = yf.Ticker("BTC-USD")
# Récupérer les 5 dernières années de données
end_date = datetime(2025, 3, 15).date()
start_date = end_date - timedelta(days=10*365)
data = btc.history(start=start_date, end=end_date)

# Afficher les premières lignes du DataFrame
print("\nAperçu des données:")
print(data.head())
print(f"\nNombre total d'observations: {len(data)}")

# Vérifier les valeurs manquantes
print("\nVérification des valeurs manquantes:")
print(data.isnull().sum())

# Nettoyer les données si nécessaire
data_cleaned = data.copy()
if data_cleaned.isnull().sum().sum() > 0:
    data_cleaned = data_cleaned.dropna()
    print(f"Données après nettoyage: {len(data_cleaned)} observations")

# Visualiser l'évolution du prix du Bitcoin
plt.figure(figsize=(14, 7))
plt.title('Prix historique du Bitcoin')
plt.plot(data_cleaned['Close'])
plt.xlabel('Date')
plt.ylabel('Prix de clôture (USD)')
plt.show()

# 2. Préparation des données pour le modèle LSTM
def prepare_data_for_lstm(data, lookback=60):
    """
    Prépare les données pour l'entraînement du modèle LSTM
    
    Args:
        data: DataFrame contenant les prix du Bitcoin
        lookback: Nombre de jours à considérer pour prédire le jour suivant
        
    Returns:
        X: Séquences d'entrée (échantillons, lookback, features)
        y: Valeurs cibles
        scaler: Scaler utilisé pour normaliser les données
    """
    # Normalisation des données
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1, 1))
    
    # Création des séquences
    X, y = [], []
    for i in range(lookback, len(scaled_data)):
        X.append(scaled_data[i-lookback:i])
        y.append(scaled_data[i])
    
    X, y = np.array(X), np.array(y)
    
    return X, y, scaler

# Paramètres
lookback = 60  # 60 jours de contexte
X, y, scaler = prepare_data_for_lstm(data_cleaned, lookback)

# Division des données en ensembles d'entraînement et de test (80/20)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(f"\nDimensions des données:")
print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"X_test: {X_test.shape}, y_test: {y_test.shape}")

# 3. Création du modèle LSTM
lstm_model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(lookback, 1)),
    Dropout(0.2),
    LSTM(units=50, return_sequences=True),
    Dropout(0.2),
    LSTM(units=50),
    Dropout(0.2),
    Dense(units=1)
])

# Compilation
lstm_model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
print(lstm_model.summary())

# 4. Entraînement
history = lstm_model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)

# 5. Prédictions
train_predict = lstm_model.predict(X_train)
test_predict = lstm_model.predict(X_test)

# Inverse scaling
train_predict = scaler.inverse_transform(train_predict)
y_train_inv = scaler.inverse_transform(y_train)
test_predict = scaler.inverse_transform(test_predict)
y_test_inv = scaler.inverse_transform(y_test)

# 6. Évaluation
# Métriques
train_rmse = np.sqrt(mean_squared_error(y_train_inv, train_predict))
test_rmse = np.sqrt(mean_squared_error(y_test_inv, test_predict))
train_r2 = r2_score(y_train_inv, train_predict)
test_r2 = r2_score(y_test_inv, test_predict)

print('\nMétriques d\'évaluation :')
print(f'Train RMSE: {train_rmse:.2f}')
print(f'Test RMSE: {test_rmse:.2f}')
print(f'Train R²: {train_r2:.4f}')
print(f'Test R²: {test_r2:.4f}')

# 7. Visualisations
# Courbe d'apprentissage
plt.figure(figsize=(12, 4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Courbe d\'apprentissage')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Prédictions vs Réalité
plt.figure(figsize=(12, 6))
# Données d'entraînement
plt.plot(data_cleaned.index[lookback:train_size+lookback], 
         y_train_inv, 
         label='Train Réel')
plt.plot(data_cleaned.index[lookback:train_size+lookback], 
         train_predict, 
         label='Train Prédictions')
# Données de test
plt.plot(data_cleaned.index[train_size+lookback:], 
         y_test_inv, 
         label='Test Réel')
plt.plot(data_cleaned.index[train_size+lookback:], 
         test_predict, 
         label='Test Prédictions')
plt.title('Prédictions LSTM vs Réalité')
plt.xlabel('Date')
plt.ylabel('Prix (USD)')
plt.legend()
plt.show()

# 8. Prédictions futures
def predict_future(lstm_model, last_sequence, n_steps=30):
    """
    Prédit les prix futurs du Bitcoin
    
    Args:
        lstm_model: Modèle LSTM entraîné
        last_sequence: Dernière séquence connue
        n_steps: Nombre de jours à prédire
        
    Returns:
        future_predictions: Array des prédictions futures
    """
    future_predictions = []
    current_sequence = last_sequence.copy()
    
    for _ in range(n_steps):
        # Prédire la prochaine valeur
        next_pred = lstm_model.predict(current_sequence.reshape(1, lookback, 1), verbose=0)
        future_predictions.append(next_pred[0, 0])
        # Mettre à jour la séquence
        current_sequence = np.roll(current_sequence, -1)
        current_sequence[-1] = next_pred
        
    return np.array(future_predictions)

# Prédire les 30 prochains jours
future_pred = predict_future(lstm_model, X[-1], 30)
future_pred = scaler.inverse_transform(future_pred.reshape(-1, 1))

# Afficher les prédictions futures
future_dates = pd.date_range(start=data_cleaned.index[-1], periods=31)[1:]
plt.figure(figsize=(12, 6))
plt.plot(data_cleaned.index[-100:], data_cleaned['Close'].values[-100:], 
         label='Historique')
plt.plot(future_dates, future_pred, label='Prédictions futures', 
         linestyle='--')
plt.title('Prédictions futures du Bitcoin avec LSTM')
plt.xlabel('Date')
plt.ylabel('Prix (USD)')
plt.legend()
plt.show()

# 9. Sauvegarde du modèle et du scaler
# Créer le dossier s'il n'existe pas
if not os.path.exists('bitcoin_model'):
    os.makedirs('bitcoin_model')

# Sauvegarder le modèle et le scaler
lstm_model.save('bitcoin_model/lstm_model.keras')
joblib.dump(scaler, 'bitcoin_model/scaler.joblib')

print("\nModèle sauvegardé dans : bitcoin_model/")
print("Fichiers exportés :")
print("- Modèle LSTM : bitcoin_model/lstm_model.keras")
print("- Scaler : bitcoin_model/scaler.joblib")

# 10. Analyse comparative avec d'autres crypto-monnaies (optionnel)
# Récupérer les données d'Ethereum pour comparaison
eth = yf.Ticker("ETH-USD")
eth_data = eth.history(start=start_date, end=end_date)

# Normaliser les deux séries pour comparaison
btc_norm = data_cleaned['Close'] / data_cleaned['Close'].iloc[0]
eth_norm = eth_data['Close'] / eth_data['Close'].iloc[0]

# Visualiser la comparaison
plt.figure(figsize=(12, 6))
plt.plot(btc_norm, label='Bitcoin')
plt.plot(eth_norm, label='Ethereum')
plt.title('Performance relative: Bitcoin vs Ethereum')
plt.xlabel('Date')
plt.ylabel('Prix normalisé (base 1)')
plt.legend()
plt.show()

# 11. Analyse de la volatilité
# Calculer les rendements journaliers
btc_returns = data_cleaned['Close'].pct_change().dropna()

# Visualiser la distribution des rendements
plt.figure(figsize=(12, 6))
plt.hist(btc_returns, bins=100, alpha=0.75)
plt.title('Distribution des rendements journaliers du Bitcoin')
plt.xlabel('Rendement journalier')
plt.ylabel('Fréquence')
plt.show()

# Calculer la volatilité sur une fenêtre glissante de 30 jours
volatility = btc_returns.rolling(window=30).std() * np.sqrt(30)  # Annualisée

plt.figure(figsize=(12, 6))
plt.plot(volatility)
plt.title('Volatilité du Bitcoin (fenêtre glissante de 30 jours)')
plt.xlabel('Date')
plt.ylabel('Volatilité')
plt.show()

print("\nAnalyse complète du Bitcoin avec LSTM terminée!")
```
